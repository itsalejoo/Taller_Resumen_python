```python
##Importaci√≥n de bibliotecas: Importamos las bibliotecas NumPy y Matplotlib para realizar c√°lculos num√©ricos y graficar, respectivamente.
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# Definici√≥n del sistema de ecuaciones: Definimos el sistema de ecuaciones como un arreglo numpy A para los coeficientes de las inc√≥gnitas y un arreglo B para los resultados de las ecuaciones.
A = np.array([[3, 9, -10], [1, -6, 4], [10, -2, 8]])
B = np.array([24, -4, 20])

# Resoluci√≥n del sistema de ecuaciones: Utilizamos la funci√≥n np.linalg.solve() de NumPy para encontrar la soluci√≥n num√©rica del sistema de ecuaciones.
solution = np.linalg.solve(A, B)

# Impresi√≥n de la soluci√≥n: Imprimimos los valores de las inc√≥gnitas (x, y, z) que resuelven el sistema de ecuaciones.
print("La soluci√≥n del sistema de ecuaciones es:")
print("x =", solution[0])
print("y =", solution[1])
print("z =", solution[2])

#Graficaci√≥n de las ecuaciones: Creamos una figura 3D y graficamos las superficies correspondientes a las ecuaciones del sistema.
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-10, 10, 100) #defino las coordenadas
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z1 = (-A[0,0]*X - A[0,1]*Y + B[0]) / A[0,2] #defino las coordenadas Z en funci√≥n de X, Y
Z2 = (-A[1,0]*X - A[1,1]*Y + B[1]) / A[1,2]
Z3 = (-A[2,0]*X - A[2,1]*Y + B[2]) / A[2,2]

ax.plot_surface(X, Y, Z1, alpha=0.5) #trazando las lineas de las coordenadas X, Y, Z1
ax.plot_surface(X, Y, Z2, alpha=0.5)
ax.plot_surface(X, Y, Z3, alpha=0.5)
# Colores para las superficies
color_map_x = cm.Reds
color_map_y = cm.Greens
color_map_z = cm.Blues

surf1 = ax.plot_surface(X, Y, Z1, cmap=color_map_x, alpha=0.5) #Se aplica un esquema de color definido por color_map_z, y la superficie se traza con una transparencia de 0.5. El objeto surf3 guarda la referencia a esta superficie.
surf2 = ax.plot_surface(X, Y, Z2, cmap=color_map_y, alpha=0.5)
surf3 = ax.plot_surface(X, Y, Z3, cmap=color_map_z, alpha=0.5)
# Agregar una barra de color
fig.colorbar(surf1, ax=ax, shrink=0.5, aspect=5)

ax.set_xlabel('x') #establecen las etiquetas de los ejes x, y, y z respectivamente
ax.set_ylabel('y')
ax.set_zlabel('z')
#En resumen, este c√≥digo resuelve un sistema de ecuaciones lineales, imprime la soluci√≥n y grafica las ecuaciones en un espacio 3D. Las ecuaciones se grafican como superficies en funci√≥n de x, y, z.
```
